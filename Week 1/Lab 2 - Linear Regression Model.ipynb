### This lab will focus on building a linear regression model. 

```python
# Tools (Numpy for computing, Matplotlib for plotting)
import numpy as np
import matplotlib.pyplot as plt
#plt.style.use("./deeplearning.mpstyle")
```

```python
# Create feature and target variables 
x_train = np.array([1.0, 2.0]) # x_train is in the input (size in 1000 square feet)
y_train = np.array([300.0, 500.0]) # y_train is the target (price in 1000s of dollars)
print(f"x_train = {x_train}")
print(f"y_train = {y_train}") 
```

```python
# m denotes the size of the training set 
# use array.shape to get the shape of the numpy array
print(f"x_train.shape: {x_train.shape}")
m = x_train.shape[0]
print(f"Number of training examples is: {m}")
```

```python
# or use pythons len() function 
m = len(x_train)
print(f"Number of training examples is {m}")
```

### Lets work with the training examples x_i, y_i 

```python
i = 0 # denotes the 1st training example
x_i = x_train[i]
y_i = y_train[i]
print(f"(x^({i}), y^({i})) = ({x_i}, {y_i})")
```

### Lets plot some data 

```python
# Plot the data points
plt.scatter(x_train, y_train, marker = "x", c = "r")
# Set title 
plt.title("Housing prices")
# Set the x-axis label 
plt.xlabel("Size (1000 sqft)")
# Set the y-axis label
plt.ylabel("Housing prices (Dollars)")
plt.show()
```

```python
# Setting parameters (b = y-intercept, w = slope) for the linear function
w = 200
b = 100
print(f"w: {w} and b: {b}")
```

```python
def compute_model_output(x, w, b): 
    """
    Computes the prediction of a linear model
    Args: 
        x (ndarray (m, )): Data, m examples
        w, b (scalar)    : model parameters
    Return
        f_wb (ndarray (m, )): model prediction
    """
    m = x.shape[0]
    f_wb = np.zeros(m) # Preallocated space / placeholder for output 
    for i in range(m): 
        f_wb[i] = w*x[i] + b
    return f_wb
```

```python
# Temporary output
tmp_f_wb = compute_model_output(x_train, w, b)

# Plot our model prediction
plt.plot(x_train, tmp_f_wb, c = "b", label = "Our Prediction")

# Plot the data poistions
plt.scatter(x_train, y_train, marker = "x", c = "r" , label = "Actual Values")

# Set the title 
plt.title("Housing prices")
# Set the y-axis label
plt.ylabel("Price in 1000s of dollars")
# Set the x-axis label 
plt.xlabel("Size  in 1000 sqft")
plt.legend()
plt.show()
```
